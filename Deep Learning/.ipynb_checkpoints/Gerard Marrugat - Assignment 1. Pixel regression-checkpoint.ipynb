# Assignment 1: Pixel Regression.

Can we recover an image by learning a deep regression map from pixels to colors?: $f: (x,y) \to (r,g,b)$.

Our target image will be Mona Lisa:

```python
import matplotlib.image as mpimg
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline

im = mpimg.imread("https://raw.githubusercontent.com/DataScienceUB/DeepLearningMaster2019/master/data/monalisa.png")

plt.imshow(im)
plt.show()
im.shape
```

Ourt training dataset will be composed of pixels locations and input and pixel values as output:

```python
X_train = []
Y_train = []
for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        X_train.append([float(i),float(j)])
        Y_train.append(im[i][j][:3])
        
X_train = np.array(X_train)
Y_train = np.array(Y_train)
print('Samples:', X_train.shape[0])
print('(x,y):', X_train[1230],'->', '(r,g,b):',Y_train[0])
```

Our objective is to train a deep Multilayer Perceptron (MLP) that is able to generate an approximate image, at least as this one:

![alt text](https://github.com/DataScienceUB/DeepLearningMaster2019/blob/master/images/result.png?raw=1)


## The Model

We are going to use Keras, a neural networks API, to build a model able to generate the Mona Lisa image.

The model is composed by nine layers; an input layer that has one hundred neurons, seven hidden layers which three of them have five hundred neurons and the rest one hundred and finally an output layer with 3 neurons. 

The activation function used between dense layers is the relu function except in the output layer where the linear function is used instead. This is because we want that the output of the model returns the exact RGB value of each pixel.

The weights of each layer are initialized using a Truncated Normal distribution with mean equal to 0.0 and standard deviation equal to 0.05 (weights random values that exceeds the double of the standard deviation from the mean are discared and redrawn). According to the Keras documentation this is the recommended initializer for neural network weights.

The chosen optimizer is adam, with a learning rate of 0.001 and a decay of 1e-5 at each update.

Mean squared error is the loss function and the accuracy the metric to be monitored during the training phase.

At the end of the next cell, the summary() instruction is executed to be able to visualize the number of parameters at each layer.

```python
import tensorflow.keras as keras
from keras.models import Sequential
from keras.layers import Dense
from keras.initializers import TruncatedNormal
from keras import optimizers


# your model here
# hints: k*10^2 neurons per layer + good initialization + deep network (>2 layers)
model = Sequential()
model.add(Dense(100,activation='relu',input_shape=(2,),kernel_initializer=TruncatedNormal(mean=0.0, stddev=0.05, seed=None)))
model.add(Dense(100,activation='relu',kernel_initializer=TruncatedNormal(mean=0.0, stddev=0.05, seed=None)))
model.add(Dense(100,activation='relu',kernel_initializer=TruncatedNormal(mean=0.0, stddev=0.05, seed=None)))
model.add(Dense(500,activation='relu',kernel_initializer=TruncatedNormal(mean=0.0, stddev=0.05, seed=None)))
model.add(Dense(500,activation='relu',kernel_initializer=TruncatedNormal(mean=0.0, stddev=0.05, seed=None)))
model.add(Dense(500,activation='relu',kernel_initializer=TruncatedNormal(mean=0.0, stddev=0.05, seed=None)))
model.add(Dense(100,activation='relu',kernel_initializer=TruncatedNormal(mean=0.0, stddev=0.05, seed=None)))
model.add(Dense(100,activation='relu',kernel_initializer=TruncatedNormal(mean=0.0, stddev=0.05, seed=None)))
model.add(Dense(3,activation='linear'))

adam = optimizers.Adam(lr=0.001, decay=1e-5)

model.compile(optimizer=adam, 
              loss='mean_squared_error',
              metrics=['accuracy'])

model.summary()
```

## Training

The model is trained with a batch size of one hundred along one thousand epochs.

```python
# use this cell to find the best model architecture

model.fit(X_train, Y_train, nb_epoch=1000, shuffle=True, verbose=1, batch_size=100)
```

## Results

At the end of the training, it is achieved a loss value of 0.0017 and an accuracy of 91,78%.

The result of the image reconstructed is the following one.

```python
Y = model.predict(X_train, batch_size=10000)
k = 0
im_out = im[:]
for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        im_out[i,j,:3]= Y[k]
        k += 1
        
plt.imshow(im_out)
plt.show()
```
